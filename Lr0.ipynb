```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = {
    'Колбаса': np.random.normal(84, 4, 12),
    'Сыр': np.random.normal(75, 5, 12),
    'Сосиски': np.random.normal(93, 3, 12),
    'Хлеб': np.random.normal(55, 5, 12),
    'Батон': np.random.normal(92, 32, 12),
    'Сардельки': np.random.normal(121, 41, 12),
    'Молоко': np.random.normal(48, 8, 12),
    'Чай': np.random.normal(71, 11, 12),
    'Кофе': np.random.normal(13, 3, 12),
    'Вода': np.random.normal(76, 16, 12),
}
```

```python
df = pd.DataFrame(data)
df
```

```python
p0 = df.sum() / df.shape[0]
p0
```

```python
std = ((df - p0) ** 2).sum() / (df.shape[0] - 1)
std = std ** (1 / 2)
std
```

```python
extrapol = p0 + np.random.normal(0, std, len(p0))
pd.concat([df, pd.DataFrame([extrapol], columns=extrapol.index)]).reset_index(drop=True)
```

```python
reliability = std / p0
reliability
```

```python
cond1 = ((df - p0) < 2 * std).all()
cond1
```

```python
cond2 = p0 > 2 * std
cond2
```

```python
cond3 = (df > 0).all()
cond3
```

```python
product_color = pd.Series(dtype='string')
for name in df.columns:
    if (not cond3[name]):
        product_color[name] = 'Красный'
    elif (not cond1[name] and not cond2[name]):
        product_color[name] = 'Оранжевый'
    elif (not cond1[name] or not cond2[name]):
        product_color[name] = 'Желтый'
    else:
        product_color[name] = 'Зеленый'
product_color
```
